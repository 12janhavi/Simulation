```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_simulations = 1000
sample_size = 100
true_mean = 10
true_std = 2
```

```python
simulated_data = np.random.normal(true_mean, true_std, size=(num_simulations, sample_size))
```

```python
# Analyze results
sample_means = np.mean(simulated_data, axis=1)
sample_stds = np.std(simulated_data, axis=1)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=30, alpha=0.75, label='Sample Means')
plt.axvline(true_mean, color='r', linestyle='dashed', linewidth=1, label='True Mean')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.title('Distribution of Sample Means')
plt.legend()
plt.grid(True)
```
